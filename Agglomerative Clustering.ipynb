```python
#======== Import Library ===============

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
# from sklearn.datasets import make_blobs
```

```python
#======== Create Dataset ===============

# dataset = make_blobs(n_samples = 200,
#                     n_features = 2,
#                     centers = 4,
#                     cluster_std = 1.6,
#                     random_state = 50)

X = []; Y = []

for i in range(50):
    X.append(np.random.randint(1,100))
    Y.append(np.random.randint(1,100))

dataset = pd.DataFrame({
    'x' : X,
    'y' : Y
})
```

```python
#======== Create Dendrogram ==============

dendrogram = sch.dendrogram(sch.linkage(dataset, method='ward'))
```

```python
#======== Plot the Data =================

plt.scatter(dataset['x'], dataset['y'], s= 100)
```

```python
#======== Perform The Actual Clustering ==========

hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage='ward')
```

```python
y_hc = hc.fit_predict(dataset)
```
